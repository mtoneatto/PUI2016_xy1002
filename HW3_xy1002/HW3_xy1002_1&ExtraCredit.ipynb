```python
from __future__ import print_function
import pylab as pl
import pandas as pd
import numpy as np

%pylab inline
```

```python
#used skeleton & instruction notebook wrote by professor
md = {}
# md is an empty disctionary
distribusions =  ['Normal','Poisson','Binomial','Chisq','Exponential']
mymean = 1;
```

```python
np.random.seed(999)
mysize = (2000 / np.array(range(1,201,2)))
[mysize,mysize.shape]
```

```python
np.random.seed(999)

#Normal Distribution:
md['Normal'] = {}
for n1 in mysize:
    md['Normal'][n1] = np.random.normal(mymean, 1, size = n1)

#Poisson Distribution:
md['Poisson'] = {}
for n2 in mysize:
    md['Poisson'][n2] = np.random.poisson(mymean, size = n2)

#Binomial Distribution:
md['Binomial'] = {}
for n3 in mysize:
    p = float(mymean)/float(n3)
    md['Binomial'][n3] = np.random.binomial(n3, p, size = n3)

#Chi-Squared Distribution:
md['Chisq'] = {}
for n4 in mysize:
    md['Chisq'][n4] = np.random.chisquare(mymean, size = n4)

#Exponential Distribution:
md['Expon'] = {}
for n5 in mysize:
    md['Expon'][n5] = np.random.exponential(1/mymean, size = n5)
```

```python

# Ploting for Normal Distribtuion:   
md['Normal']['means'] = {}
axnormal = pl.figure(figsize=(10,30)).add_subplot(511)

for nn1 in md['Normal']:
    if not type(nn1) == str:
        md['Normal']['means'][nn1] = md['Normal'][nn1].mean()
        #and plot it
        axnormal.plot(nn1, md['Normal']['means'][nn1], 'o')
        axnormal.set_xlabel('sample size', fontsize=18)
        axnormal.set_ylabel('sample mean', fontsize=18)
        axnormal.set_title('Normal Distribution', fontsize=18)
        axnormal.plot([min(mysize), max(mysize)], [mymean, mymean], 'k')
        axnormal.text(0,0.3,'Figure 1: For 100 Normal Distribution we can see that when the ' +
                    'sample size increases, \n the sample means tend to be more close to the population mean.',
                      fontsize = 12)
        
# Ploting for Poisson Distribtuion: 
md['Poisson']['means'] = {}
axpoisson = pl.figure(figsize=(10,30)).add_subplot(512)

for nn2 in md['Poisson']:
    if not type(nn2) == str:
        md['Poisson']['means'][nn2] = md['Poisson'][nn2].mean()
        #and plot it
        axpoisson.plot(nn2, md['Poisson']['means'][nn2], 'o')
        axpoisson.set_xlabel('sample size', fontsize=18)
        axpoisson.set_ylabel('sample mean', fontsize=18)
        axpoisson.set_title('Poisson Distribution', fontsize=18)
        axpoisson.plot([min(mysize), max(mysize)], [mymean, mymean], 'k')
        axpoisson.text(0,0.4,'Figure 2: For 100 Poisson Distribution we can see that when the ' +
                    'sample size increases, \n the sample means tend to be more close to the population mean.',
                      fontsize = 12)
        
# Ploting for Binomial Distribtuion:         
md['Binomial']['means'] = {}
axbinomial = pl.figure(figsize=(10,30)).add_subplot(513)

for nn3 in md['Binomial']:
    if not type(nn3) == str:
        md['Binomial']['means'][nn3] = md['Binomial'][nn3].mean()
        #and plot it
        axbinomial.plot(nn3, md['Binomial']['means'][nn3], 'o')
        axbinomial.set_xlabel('sample size', fontsize=18)
        axbinomial.set_ylabel('sample mean', fontsize=18)
        axbinomial.set_title('Binomial Distribution', fontsize=18)
        axbinomial.plot([min(mysize), max(mysize)], [mymean, mymean], 'k')
        axbinomial.text(0,0.5,'Figure 3: For 100 Binomial Distribution we can see that when the ' +
                    'sample size increases, \n the sample means tend to be more close to the population mean.',
                    fontsize = 12)
 
# Ploting for Chi-Squared Distribtuion:         
md['Chisq']['means'] = {}
axchisq = pl.figure(figsize=(10,30)).add_subplot(514)

for nn4 in md['Chisq']:
    if not type(nn4) == str:
        md['Chisq']['means'][nn4] = md['Chisq'][nn4].mean()
        #and plot it
        axchisq.plot(nn4, md['Chisq']['means'][nn4], 'o')
        axchisq.set_xlabel('sample size', fontsize=18)
        axchisq.set_ylabel('sample mean', fontsize=18)
        axchisq.set_title('Chi-Squared Distribution', fontsize=18)
        axchisq.plot([min(mysize), max(mysize)], [mymean, mymean], 'k')
        axchisq.text(0,0,'Figure 4: For 100 Chi-Squared Distribution we can see that when the ' +
                    'sample size increases, \n the sample means tend to be more close to the population mean.',
                    fontsize = 12)

# Ploting for Exponential Distribtuion:         
md['Expon']['means'] = {}
axexpon = pl.figure(figsize=(10,30)).add_subplot(515)

for nn5 in md['Expon']:
    if not type(nn5) == str:
        md['Expon']['means'][nn5] = md['Expon'][nn5].mean()
        #and plot it
        axexpon.plot(nn5, md['Expon']['means'][nn5], 'o')
        axexpon.set_xlabel('sample size', fontsize=18)
        axexpon.set_ylabel('sample mean', fontsize=18)
        axexpon.set_title('Exponential Distribution', fontsize=18)
        axexpon.plot([min(mysize), max(mysize)], [mymean, mymean], 'k')
        axexpon.text(0,0.4,'Figure 5: For 100 Exponential Distribution we can see that when the ' +
                    'sample size increases, \n the sample means tend to be more close to the population mean.',
                    fontsize = 12)

```

```python
## Extra Credit:
## Reference: http://stackoverflow.com/questions/20011122/fitting-a-normal-distribution-to-1d-data
from scipy.stats import norm
```

```python
## Fit for Hitogram of Normal Distribution Means
Normal_allmeans = list(md['Normal']['means'].values())

axNormal_allmeans = pl.figure(figsize=(6, 6)).add_subplot(111)
axNormal_allmeans.hist(Normal_allmeans,bins=50)
axNormal_allmeans.set_title('1.Fit Result: Normal Distribution Means', fontsize=18)
axNormal_allmeans.set_xlabel('sample mean', fontsize = 18)
axNormal_allmeans.set_ylabel('N', fontsize = 18)


xmin, xmax = axNormal_allmeans.set_xlim()
x = np.linspace(xmin, xmax, 100)
m = np.mean(Normal_allmeans)
s = np.std(Normal_allmeans)
axNormal_allmeans.plot(x, norm.pdf(x,m,s),'k', linewidth=2)
```
